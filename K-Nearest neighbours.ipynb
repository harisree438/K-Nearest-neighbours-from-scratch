```python
import math
import csv
import numpy as np

# Read the CSV file and store its rows in a list
data_list = []
with open('wine.data', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        data_list.append(row)

# Convert the list of rows to a NumPy array
data = np.array(data_list)

def euclidean_distance(row1, row2):
    # Initialize the distance variable
    distance = 0

    # Calculate the squared differences and accumulate them
    for i in range(1, (len(row1)-1)):
        distance += (float(row1[i]) - float(row2[i])) ** 2


    # Calculate the square root of the accumulated squared differences
    distance = math.sqrt(distance)
    return distance
row0= data[len(data)-1]

dist=list()
for row in data:
  distance = euclidean_distance(row0, row)
  dist.append(distance)
print(distance)
print(np.shape(distance))
print(dist)
print(len(dist))
```
